```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# BOSTON HOUSING DATASET

# 1. Load the dataset
df = pd.read_csv('../_raw/regr_housing.csv')

# 2. Drop high-cardinality and spatial identifier columns
df_clean = df.drop(columns=['TOWN', 'TRACT', 'LON', 'LAT'])

# 3. Rename columns to be readable and descriptive
column_mapping = {
    'CRIM': 'Crime_Rate',
    'ZN': 'Zoned_Land_Proportion',
    'INDUS': 'Non_Retail_Business_Proportion',
    'CHAS': 'Charles_River_Proximity',
    'NOX': 'Nitric_Oxide_Concentration',
    'RM': 'Average_Rooms',
    'AGE': 'Property_Age',
    'DIS': 'Distance_to_Employment',
    'RAD': 'Highway_Accessibility_Index',
    'TAX': 'Property_Tax_Rate',
    'PTRATIO': 'Pupil_Teacher_Ratio',
    'MEDV': 'Median_Home_Value'
}
df_clean = df_clean.rename(columns=column_mapping)

# 4. Separate features (X) and target (y)
X = df_clean.drop(columns=['Median_Home_Value'])
y = df_clean['Median_Home_Value']

# 5. Normalize the features (Mean=0, StdDev=1)
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# 6. Recombine the normalized features with the target variable
df_final = pd.concat([X_scaled, y], axis=1)

# 7. Export to a clean CSV
df_final.to_csv('cleaned_normalized_housing.csv', index=False)

print("Process complete! Saved as 'cleaned_normalized_housing.csv'")
print(f"Final shape: {df_final.shape}")
print("\nNew readable columns:\n", df_final.columns.tolist())
```

```python
# BREAST CANCER DATASET

# 1. Load the dataset
df = pd.read_csv('../_raw/bin_breast_cancer.csv')

# 2. Drop the ID column AND the empty trailing column
# errors='ignore' ensures it won't crash if the 'Unnamed: 32' column is missing
df_clean = df.drop(columns=['id', 'Unnamed: 32'], errors='ignore')

# 3. Encode the target variable (Malignant = 1, Benign = 0)
df_clean['diagnosis'] = df_clean['diagnosis'].map({'M': 1, 'B': 0})

# 4. Make column names clean and uniform (Title_Case)
df_clean.columns = [col.replace('_', ' ').title().replace(' ', '_') for col in df_clean.columns]

# 5. Separate features (X) and target (y)
X = df_clean.drop(columns=['Diagnosis'])
y = df_clean['Diagnosis']

# 6. Normalize the features (Mean=0, StdDev=1)
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# 7. Recombine the normalized features with the encoded target variable
df_final = pd.concat([X_scaled, y], axis=1)

# 8. Export to a clean CSV
df_final.to_csv('cleaned_normalized_breast_cancer.csv', index=False)

print("Process complete! Saved as 'cleaned_normalized_breast_cancer.csv'")
print(f"Final shape: {df_final.shape} (The ghost column is gone!)")
```

```python
# IRIS DATASET

# 1. Load the dataset
df = pd.read_csv('../_raw/mult_iris.csv')

# 2. Drop the irrelevant identifier column (errors='ignore' for safety)
df_clean = df.drop(columns=['Id', 'Unnamed: 6'], errors='ignore')

# 3. Rename columns to be cleaner (removing the 'Cm' suffix)
column_mapping = {
    'SepalLengthCm': 'Sepal_Length',
    'SepalWidthCm':  'Sepal_Width',
    'PetalLengthCm': 'Petal_Length',
    'PetalWidthCm':  'Petal_Width'
}
df_clean = df_clean.rename(columns=column_mapping)

# 4. Encode the target variable for multiclass (0, 1, 2)
species_mapping = {
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
}
df_clean['Species'] = df_clean['Species'].map(species_mapping)

# 5. Separate features (X) and target (y)
X = df_clean.drop(columns=['Species'])
y = df_clean['Species']

# 6. Normalize the features (Mean=0, StdDev=1)
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# 7. Recombine the normalized features with the encoded target variable
df_final = pd.concat([X_scaled, y], axis=1)

# 8. Export to a clean CSV
df_final.to_csv('cleaned_normalized_iris.csv', index=False)

print("Process complete! Saved as 'cleaned_normalized_iris.csv'")
print(f"Final shape: {df_final.shape}")
print("\nTarget variable (Species) distribution:")
print(df_final['Species'].value_counts().sort_index())
```
